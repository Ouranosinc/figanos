```python
#import
import xarray as xr
%matplotlib inline
%config InlineBackend.print_figure_kwargs = {'bbox_inches':'tight'}
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import cartopy.crs as ccrs

import spirograph.matplotlib as sp
```

```python
# use ouranos style
sp.utils.set_mpl_style('ouranos')
```

```python
print(sp.gdfmap.__doc__)
```

```python
rivers = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_10m_rivers_north_america.geojson')
```

```python
#projection is correct but earth is cutoff. This is specifying a projection to the matplotlib axes and using df.plot(ax=ax)
ax = sp.gdfmap(rivers, 'scalerank', 
               projection = ccrs.LambertConformal(),
               features = {'ocean': {'color':'#a2bdeb'}},
               cbar_kw = {'orientation': 'vertical'},
               frame=True
              )
```

```python
bassins = gpd.read_file('https://stqc380donopppdtce01.blob.core.windows.net/donnees-ouvertes/Zones_gestion_integree_eau/ZGIEBV.json')
```

```python
sp.gdfmap(bassins, 
          'ZGIE_KM2',
          cmap = 'viridis', 
          cmap_bounds = [0,10000,100000,1000000], 
          cbar_kw = {'label': 'Watershed Area (km2)'})
```
